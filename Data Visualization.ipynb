# Data Visualization

This file imports the functionalities from 'data_visualization_module.py' which are used for generating the plots for the Visualization of the FIFA Player Attributes. The plots generated include the 'Categorical Attributes Radar Plots', which are radar plots of attributes for each player category, 'Comprehensive Attributes Radar Plot' which is the radar plot comparing the attributes of the strikers, midfielders and defenders.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import subplots
from math import pi

from data_visualization_module import *
```

### Lists of Player Attributes, Titles and Number of Players used in the 'Categorical Attributes Radar Plots'

```python
# List of Attributes
Attacking = ['Dribbling','Volleys','Finishing','Short Passing','Heading Accuracy']
MovementAndPower = ['Sprint Speed','Reactions','Positioning','Agility','Shot Power','Stamina']

Midfielding = ['Short Passing','Dribbling','Ball Control','Long Passing','Crossing']
PowerAndMentality = ['Long Shots','Stamina','Balance','Strength','Vision']

Defending = ['Marking','Standing Tackle','Sliding Tackle','Strength','Interceptions']

GoalKeeping = ['GK Diving','GK Handling','GK Kicking','GK Positioning','GK Reflexes']

attributes_by_category = {'Striker': [Attacking, MovementAndPower],
                          'Midfielder': [Midfielding, PowerAndMentality],
                          'Defender': [Defending],
                          'GoalKeeper': [GoalKeeping]}

# Parts of Titles for the Radar Plots 
title_parts = ['Attacking', 'Movement and Power', 'Midfielding','Power and Mentality','Defending','GoalKeeping']

# Number of Players by Category and Percentage
percentage = [5,20,50,100]
no_of_players_by_category = {'Striker': 2000, 'Midfielder': 4000, 'Defender': 3000, 'GoalKeeper': 1000}
no_of_players_by_percentage = {'Striker': [], 'Midfielder': [], 'Defender': [], 'GoalKeeper': []}

for category, number in no_of_players_by_category.items(): 
    no_of_players_by_percentage[category] = [int(p*number/100) for p in percentage]
```

### Generation of the 'Categorical Attributes Radar Plots'

```python
plot_count = 0
for category, attr_list in attributes_by_category.items():
    for attributes in attr_list:
        title_part = title_parts[plot_count]
        plot_count += 1
        categorical_attributes_radar_plot(category, attributes, percentage, no_of_players_by_percentage[category], title_part)
```

### List of Attributes and Top Number of Players used in the 'Comprehensive Attributes Radar Plot'

```python
comparison_attr_dict = {'Reactions':0, 'Passing':0, 'Sprint Speed':0, 'Finishing':0, 'Volleys':0,
                        'Strength':0, 'Stamina':0, 'Tackling':0, 'Interceptions':0}
comparison_attr_list = ['Reactions', 'Short Passing', 'Long Passing', 'Sprint Speed', 'Finishing', 'Volleys',
                        'Strength', 'Stamina', 'Standing Tackle', 'Sliding Tackle', 'Interceptions']
comparison_attr_map  = {'Passing':['Short Passing', 'Long Passing'], 'Tackling':['Standing Tackle', 'Sliding Tackle']}
number_top = 100
```

### Generation of the 'Comprehensive Attributes Radar Plot'

```python
comprehensive_attributes_radar_plot(comparison_attr_dict, comparison_attr_list, comparison_attr_map, number_top)
```
