```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import figure
from math import pi

alldata = pd.read_csv('player_attributes.csv')
#define new means of all players and sorting by player location
alldata.loc['mean'] = alldata.mean()
alldata.loc['mean_Striker'] = alldata[alldata['Player Category']=='Striker'].mean()
alldata.loc['mean_Midfielder'] = alldata[alldata['Player Category']=='Midfielder'].mean()
alldata.loc['mean_Defender'] = alldata[alldata['Player Category']=='Defender'].mean()
alldata.loc['mean_GoalKeeper'] = alldata[alldata['Player Category']=='GoalKeeper'].mean()
alldata
```

```python
# radar plot for mean all players in Attack 
# all the sub-characters
attack_labels = np.array(['Crossing','Finishing','Heading Accuracy','Short Passing','Volleys'])
skill_labels = np.array(['Dribbling','Curve','FK Accuracy','Long Passing','Ball Control'])
movement_lables = np.array(['Acceleration','Sprint Speed','Agility','Reactions','Balance'])
power_lables = np.array(['Shot Power','Jumping','Stamina','Strength','Long Shots'])
mentality_lables = np.array(['Aggression','Interceptions','Positioning','Vision','Penalties','Composure'])
defending_labels = np.array(['Marking','Standing Tackle','Sliding Tackle'])
goalkeeping_lables = np.array(['GK Diving','GK Handling','GK Kicking','GK Reflexes'])
characterlist = [attack_labels,skill_labels,movement_lables,power_lables,mentality_lables,defending_labels,goalkeeping_lables]
for elem in characterlist:
    labels = elem
    print(elem)
    stats = alldata.loc['mean', labels].values
    angles = np.linspace(0,2*np.pi,len(labels),endpoint = False)
    stats = np.concatenate((stats,[stats[0]]))
    angles = np.concatenate((angles,[angles[0]]))
    fig = plt.figure()
    ax = fig.add_subplot(111,polar =True) #ser polar axis
    ax.plot(angles, stats, 'o-',linewidth = 2) # draw the plot
    ax.fill(angles, stats, alpha = 0.25) #fill the area
    ax.set_thetagrids(angles*180/np.pi, labels)
    ax.set_title('mean'+str(elem)+'of all playes')
    ax.grid(True)
```
